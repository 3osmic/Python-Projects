### Change various characteristics of a plot 

Increase the size, change plot line colors, add point markers, etc.

#### Change the figure size

If a plot appears too crowded, you can increase the size of the plot using __figsize()__

__figsize(x,y)__ takes two parameters: width and height

__Syntax:__

    -plt.figure(figsize=(x,y)) 
    
REF: https://www.geeksforgeeks.org/change-plot-size-in-matplotlib-python/


#### Add markers at the plot points

You can plot x,y value points instead of lines using the following syntax:

    plt.plot(xpoints,ypoints,marker)
    
There are many options for markers, here are just a few:
    
    markers = ('o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X')
    
w3schools.com REF: https://www.w3schools.com/python/matplotlib_markers.asp


#### Add grid lines to the plot
To make it easier to determine the exact values plotted, grid lines are often useful. 

You can add horizontal, vertical or both to a graph using __grid()__ 

- REF: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html
- REF: https://www.w3schools.com/python/matplotlib_grid.asp

## 6. PIE Charts (10 pts)

The <b>DF_SUPERBOWL</b> DataFrame has a column 'Network' which indicates which Network televised the Superbowl each year.

The list of Networks includes: __CBS, NBC, ABC, Fox__

For this exercise, you will need to "cast" a Pandas Series to a Python List. 
- This is easily done using __list()__ For example: MY_LIST = __list__(MY_SERIES) 

Python LISTs have a __.count()__ method that returns the number of times a specific value appears in the list.

You will be asked to use the __.count()__ method to get the number of times each Network value appears in a list.
- REF: https://www.w3schools.com/python/ref_list_count.asp

To generate a PIE CHART, you will need to create a few lists and count of the number of entries for each network.<br><br>
<span style="color:blue">
    1. Create a Python LIST named <b>NETWORKS</b> containing the following network abbreviations: ['CBS', 'NBC', 'ABC', 'Fox'] <br>
    2. Create a Python LIST named <b>GAMES</b> with 4 values initialized to zero: [0,0,0,0] <br>
    3. Create a Python LIST named <b>NETWORK_LIST</b> containing all the values from <b>DF_SUPERBOWL['Network']</b><br>
    - HINT: Access the column <b>DF_SUPERBOWL['Network']</b> and "cast" it to a Python LIST <br>
    4. Update the value of <b>GAMES[0]</b> with the number of occurences of <b>'CBS'</b><br>
    5. Update the value of <b>GAMES[1]</b> with the number of occurences of <b>'NBC'</b><br>
    6. Update the value of <b>GAMES[2]</b> with the number of occurences of <b>'ABC'</b><br>
    7. Update the value of <b>GAMES[3]</b> with the number of occurences of <b>'Fox'</b><br>
    8. Create a PIE Chart using the data in <b>GAMES</b> and set the labels to the values in <b>NETWORKS</b>.<br>
    9. Add the title <b>Networks that televised the Superbowl</b> <br>
    10. <strong>Show</strong> the plot.
</span>

```python
# INSERT CODE FOR STEPS 1 - 10
NETWORKS = ['CBS', 'NBC', 'ABC', 'Fox']
GAMES = [0, 0, 0, 0]
NETWORK_LIST = list(DF_SUPERBOWL['Network'])
GAMES = DF_SUPERBOWL['Network'].value_counts().reindex(NETWORKS, fill_value=0).tolist()
plt.pie(GAMES, labels=NETWORKS, autopct='%1.1f%%', startangle=90, colors=['blue', 'orange', 'green', 'red'])
plt.title('Networks that televised the Superbowl')
plt.show()
```
